```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('results (6).csv')
```

```python
data.head()
```

```python
# function for moving average
# for global_avg_temp
def moving_avg(mA_range, data_input):
    output = data_input.rolling(window = mA_range, on = "global_average_temp").mean().dropna()
    return output
# Function Calling with the range of Moving Average
mA_value = 5
chart_moving_avg = moving_avg(mA_value, data)

# Drawing the graph: Global Temperature
plt.plot(chart_moving_avg ['year'], chart_moving_avg ['global_average_temp'], label = 'Global average temp')
plt.legend()
plt.xlabel ("Years")
plt.ylabel ("Temperature (Â°C)")
plt.title ("Global Average Temperature")
plt.show ()
```

```python
# for city_avg_temp
def moving_avg(mA_range, data_input):
    output = data_input.rolling(window = mA_range, on = "city_average_temp").mean().dropna()
    return output
# Function Calling with the range of Moving Average
mA_value = 14
chart_moving_avg = moving_avg(mA_value, data)

# Drawing the graph: Global Temperature
plt.plot(chart_moving_avg ['year'], chart_moving_avg ['city_average_temp'], label = 'City - New York',color='g')
plt.legend()
plt.xlabel ("Years")
plt.ylabel ("Temperature (Â°C)")
plt.title ("City Average Temperature")
plt.show ()
```
